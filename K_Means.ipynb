```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
X= pd.read_csv("D:\Bharat\Personal\EVA6\Session10\Coco_Test.csv",usecols=["Norm_W_b","Norm_H_b"])
Y=np.array(X.values.tolist())
#X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

plt.scatter(Y[:,0], Y[:,1])
```

```python
kmeans = KMeans(n_clusters=6, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(Y)
plt.scatter(Y[:,0], Y[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')

def addlabels(x,y):
    
    for i in range(len(x)):
        str1= str(round(x[i],2)) + ', ' + str(round(y[i],2))
        plt.text(x[i],y[i],str1,fontsize="12",color="white",weight="bold")
        
addlabels(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1])
plt.show()
```
